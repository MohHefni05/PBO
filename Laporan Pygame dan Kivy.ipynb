```python
import pygame
import random


# Menentukan Warna
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
GREEN = (0, 255, 0)
RED = (255, 0, 0)

pygame.init()


# Mengatur lebar dan tinggi layar [lebar, tinggi] dan objek maupun lagu
size = (700, 500)
screen = pygame.display.set_mode(size)


""" Ini adalah kelas Bola sederhana untuk merespons bola dalam permainan."""

class Ball(object):
    def __init__ (self, screen, radius,x,y):
        self.__screen = screen
        self._radius = radius
        self._xLoc = x
        self._yLoc = y
        self.__xVel = 5
        self.__yVel = -3
        w, h = pygame.display.get_surface().get_size()
        self.__width = w
        self.__height = h
        
    def draw(self):
        """ menggambar bola ke layar. """
        pygame.draw.circle(screen,(255, 0, 0) , (self._xLoc,self._yLoc), self._radius)

    def update(self, paddle, brickwall):
        """ menggerakkan bola di layar.
            mengandung beberapa deteksi tabrakan."""
        self._xLoc += self.__xVel
        self._yLoc += self.__yVel
        if self._xLoc <= self._radius:
            self.__xVel *= -1
        elif self._xLoc >= self.__width - self._radius:
            self.__xVel *= -1
        if self._yLoc <= self._radius:
            self.__yVel *= -1
        elif self._yLoc >= self.__height - self._radius:
            return True

        # untuk memantul dari batu bata.
        if brickwall.collide(self):
            self.__yVel *= -1

        # kelenturan tabrakan antara bola dan dayung
        paddleX = paddle._xLoc
        paddleY = paddle._yLoc
        paddleW = paddle._width
        paddleH = paddle._height
        ballX = self._xLoc
        ballY = self._yLoc
        if ((ballX + self._radius) >= paddleX and ballX <= (paddleX + paddleW)) \
        and ((ballY + self._radius) >= paddleY and ballY <= (paddleY + paddleH)):
            self.__yVel *= -1
        return False        

    """ Kelas sederhana untuk mewakili pemukul """        
class Paddle (object):
    def __init__ (self, screen, width, height,x,y):
        self.__screen = screen
        self._width = width
        self._height = height
        self._xLoc = x
        self._yLoc = y
        w, h = pygame.display.get_surface().get_size()
        self.__W = w
        self.__H = h
    def draw(self):
        """ menggambar dayung ke layar. """
        pygame.draw.rect(screen, (0,0,0), (self._xLoc,self._yLoc,self._width,self._height),0)
        
    def update(self):
        """ menggerakkan pemukul di layar melalui mouse """
        x,y = pygame.mouse.get_pos()
        if x >= 0 and x <= (self.__W - self._width):
            self._xLoc = x
 
""" Kelas ini mewakili kelas Bata sederhana.
    Untuk mewakili batu bata ke layar. """
class Brick (pygame.sprite.Sprite):    
    def __init__(self, screen, width, height, x,y):
        self.__screen = screen
        self._width = width
        self._height = height
        self._xLoc = x
        self._yLoc = y
        w, h = pygame.display.get_surface().get_size()
        self.__W = w
        self.__H = h
        self.__isInGroup = False

    def draw(self):
        """ menggambar bata ke layar.
            warna: rgb(56, 177, 237) """
        pygame.draw.rect(screen, (56, 177, 237), (self._xLoc,self._yLoc,self._width,self._height),0)
        
    def add (self, group):
        """ menambahkan bata ini ke grup tertentu. """
        group.add(self)
        self.__isInGroup = True
        
    def remove(self, group):
        """  menghapus bata ini dari grup yang diberikan. """
        group.remove(self)
        self.__isInGroup = False
        
    def alive(self):
        """ mengembalikan true ketika batu bata ini adalah milik dinding bata. jika tidak salah """
        return self.__isInGroup

    def collide(self, ball):
        """ tabrakan antara bola dan batu bata ini """
        brickX = self._xLoc
        brickY = self._yLoc
        brickW = self._width
        brickH = self._height
        ballX = ball._xLoc
        ballY = ball._yLoc
        radius = ball._radius

        if ((ballX + radius) >= brickX and ballX <= (brickX + brickW)) \
        and ((ballY + radius) >= brickY and ballY <= (brickY + brickH)):
            return True
        return False

""" Ini adalah kelas sederhana untuk mewakili a dinding bata. """
class BrickWall (pygame.sprite.Group):
    def __init__ (self,screen, x, y, width, height):
        self.__screen = screen
        self._x = x
        self._y = y
        self._width = width
        self._height = height
        self._bricks = []

        X = x
        Y = y
        for i in range(3):
            for j in range(4):
                self._bricks.append(Brick(screen,width,height,X,Y))
                X += width + (width/ 7.0)
            Y += height + (height / 7.0)
            X = x
        
    def add(self,brick):
        """ menambahkan bata ke BrickWall ini (grup) """
        self._bricks.append(brick)
    def remove(self,brick):
        """ menghapus bata dari BrickWall ini (grup) """
        self._bricks.remove(brick)
        
    def draw(self):
        """ menggambar semua batu bata ke layar. """
        for brick in self._bricks:
            if brick != None:
                brick.draw()
    def update(self, ball):
        
        """ memeriksa tabrakan antara bola dan batu bata. """
        for i in range(len(self._bricks)):
            if ((self._bricks[i] != None) and self._bricks[i].collide(ball)):
                self._bricks[i] = None
        
        # removes the None-elements from the brick list.
        for brick in self._bricks:
            if brick == None:
                self._bricks.remove(brick)
    def hasWin(self):
        """ Apakah pemain memenangkan permainan? """
        return len(self._bricks) == 0

    def collide (self, ball):
        """ periksa tabrakan antara bola dan
            salah satu dari batu bata."""
        for brick in self._bricks:
            if brick.collide(ball):
                return True
        return False

# Game ini menolak bola, dayung, dan dinding bata
ball = Ball(screen,25,350,250)
paddle = Paddle(screen,200,20,250,450)
brickWall = BrickWall(screen,25,25,150,50)

pygame.display.set_caption("Pantulan Bola")
 
# Loop sampai pengguna mengklik tombol tutup.
done = False
 
# Digunakan untuk mengatur seberapa cepat layar diperbarui
clock = pygame.time.Clock()

# untuk menampilkan teks dalam game 
pygame.font.init() # Anda harus memanggil ini di awal,
                   # jika Anda ingin menggunakan modul ini.

# pesan untuk Kalah
mgGameOver = pygame.font.SysFont('Times New Roman', 40)

# pesan untuk Menang
mgWin = pygame.font.SysFont('Times New Roman', 40)


textsurfaceGameOver = mgGameOver.render('Game Habis, Kamu Kalah! wkwkwkwk', False, (255, 0, 0))
textsurfaceWin = mgWin.render("Kamu Menang! Selamat",False,(0,255,0))

isGameOver = False # menentukan apakah game kalah
gameStatus = True # game masih berjalan

score = 0 # skor untuk permainan.
 
# -------- Main Program Loop -----------
while not done:
    # --- Main event loop
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True
 
    # --- Logika permainan seharusnya ada di sini
 
    # --- Kode kliring layar ada di sini
 
    # Di sini, kami menghapus layar menjadi putih. Jangan letakkan perintah menggambar lainnya
    # di atas ini, atau mereka akan terhapus dengan perintah ini.
 
    # Jika Anda menginginkan gambar latar belakang, ganti yang jelas ini dengan blit'ing
    # gambar latar belakang.
    screen.fill(WHITE)
 
    # --- Kode menggambar harus di sini
    
    """ Karena saya menggunakan OOP dalam logika game dan kode gambar,
        keduanya di bagian yang sama """
    if gameStatus:

        # pertama menggambar bola untuk menampilkan skor dengan tepat
        brickWall.draw()

        # setelah mencetak gol. karena batu bata hit dihapus dalam metode pembaruan
        brickWall.update(ball)

        paddle.draw()
        paddle.update()

        if ball.update(paddle, brickWall):
            isGameOver = True
            gameStatus = False
            
        if brickWall.hasWin():
            gameStatus = False

        ball.draw()

    else: # game tidak berjalan.
        if isGameOver: # player kalah
            screen.blit(textsurfaceGameOver,(30,200))
        elif brickWall.hasWin(): # player menang
            screen.blit(textsurfaceWin,(70,200))

            
    pygame.display.flip()
    clock.tick(60)
pygame.quit()
```

<b> pengertian kivy </b><br>
Kivy ini sendiri merupakan framework yang dibangun menggunakan library dari bahasa pemrograman Python yang bersifat Open Source. Tujuan dikembangkannya framework ini agar dapat membantu developer secara cepat dalam mengembangkan aplikasi yang memiliki tampilan antarmuka inovatif seperti aplikasi yang mendukung multitouch.<br>
Framework ini sendiri dapat dijalankan pada sistem operasi Windows, Mac OS X, Linux, Android, hingga Raspberry Pi. Selain itu, Kivy mendukung untuk input beragam perangkat seperti WM_Touch, WM_Pen, Mac OS X Trackpad, Magi Mouse, Mtdev, Linux Kernel HID, dan TUIO.<br>

```python
import kivy
from kivy.app import App
from kivy.uix.label import Label
from kivy.uix.gridlayout import GridLayout  # one of many layout structures
from kivy.uix.textinput import TextInput  # allow for ...text input.

kivy.require("1.10.1")

# An actual app is likely to consist of many different
# "pages" or "screens." Inherit from GridLayout
class ConnectPage(GridLayout):
    # runs on initialization
    def __init__(self, **kwargs):
        # we want to run __init__ of both ConnectPage AAAAND GridLayout (understanding_super.py)
        super().__init__(**kwargs)

        self.cols = 2  # used for our grid

        # widgets added in order, so mind the order.
        self.add_widget(Label(text='IP:'))  # widget #1, top left
        self.ip = TextInput(multiline=False)  # defining self.ip...
        self.add_widget(self.ip) # widget #2, top right

        self.add_widget(Label(text='Port:'))
        self.port = TextInput(multiline=False)
        self.add_widget(self.port)

        self.add_widget(Label(text='Username:'))
        self.username = TextInput(multiline=False)
        self.add_widget(self.username)


class EpicApp(App):
    def build(self):
        return ConnectPage()


if __name__ == "__main__":
    EpicApp().run()
```

```python
from kivy.app import App
from kivy.lang import Builder
from kivy.uix.screenmanager import ScreenManager, Screen
from kivy.properties import ObjectProperty
from kivy.uix.popup import Popup
from kivy.uix.label import Label
from database import DataBase
from kivy.core.window import Window
from kivy.uix.boxlayout import BoxLayout


class BuatSettingan(Screen,BoxLayout):
    def __init__(self,**kwargs):
        super(BuatSettingan,self).__init__(**kwargs)
        ip = ObjectProperty(None)
        port = ObjectProperty(None)
        username = ObjectProperty(None)

    def tambah(self):
        if self.ip.text != "" and self.port.text != "" and self.username.text != "":
            if self.port != "":
                db.add_settingan(self.ip.text, self.port.text, self.username.text)
                self.reset()
                ConnectPage.current = "main"
            else:
                PopupGagalTambah()
        else:
            PopupGagalTambah()

    def reset(self):
        self.ip.text = ""
        self.port.text = ""
        self.username.text = ""
        
    def tutup(self):
        App.get_running_app().stop()
        Window.close()

class TampilanTambah(Screen):

    def Kembali(self):
        ConnectPage.current = "create"
        
class WindowManager(ScreenManager):
    pass

class tentang(Screen):

    def Kembali(self):
        ConnectPage.current = "create"

def PopupGagalTambah():
    pop = Popup(title='Peringatan',
                  content=Label(text='Form Tidak Boleh Kosong'),
                  size_hint=(None, None), size=(400, 400))

    pop.open()


kv = Builder.load_file("tampilan.kv")
ConnectPage = WindowManager()
db = DataBase("Data Settingan.txt")

screens = [BuatSettingan(name="create"),TampilanTambah(name="main"),tentang(name="ta")]
for screen in screens:
    ConnectPage.add_widget(screen)

ConnectPage.current = "create"


class MyMainApp(App):
    def build(self):
        return ConnectPage
    
if __name__ == "__main__":
    MyMainApp().run()
```
